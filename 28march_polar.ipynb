```python
def coeff():
    for i in range(1,n):
        for j in range(1,m):
            A[i,j]=2.0*((1/(h*h))+1/((r[i]*r[i])*(k*k)))
            B[i,j]=(1/(h*h))-(1/(r[i]*2*h))
            C[i,j]=(1/(h*h))+(1/(r[i]*2*h))
            D[i,j]=1.0/((r[i]*r[i])*(k*k))
            E[i,j]=1.0/((r[i]*r[i])*(k*k))
```

```python
def gauss():
    for k in range(1,p):
        for i in range(1,n):
            for j in range(1,m):
                U[i,j]=(B[i,j]*U[i-1,j]+C[i,j]*U[i+1,j]+D[i,j]*U[i,j-1]+E[i,j]*U[i,j+1])/A[i,j]
```

```python
def main():
    import numpy as np
    from numpy import sin , pi, sinh
    import seaborn as sns
    from numpy import sin , pi, sinh
    import matplotlib.pyplot as plt
    from mpl_toolkits import mplot3d
    global A,B,C,D,E,U,U1,n,m,p,p1,h,k,U2,r
    n=60
    m=60
    p=2
    a=2.0
    #r=1.0
    theta=pi/2
    h=a/n #delta r
    k=theta/m  #delta theta
    theta0=0 
    r0=0
    p1=10000
    eps=0.001
    R1=np.zeros(p1)
    A=np.zeros((n,m))
    B=np.zeros((n,m))
    C=np.zeros((n,m))
    D=np.zeros((n,m))
    E=np.zeros((n,m))
    U=np.zeros((n+1,m+1))
    r=np.zeros(n+1)
    theta=np.zeros(m+1)
    U1=np.zeros((n+1,m+1))
    U2=np.zeros((n+1,m+1))
    R=np.zeros((p1))
    alp=0.3
    r[0]=r0
    for i in range(1,n+1):
        r[i]=r[i-1]+h
    #print("r=",r)
    theta[0]=theta0
    for j in range(1,m+1):
        theta[j]=theta[j-1]+k
    #print("theta=",theta)
    
    for i in range(0,n):
        U[i,0]=0
        #U[i,m]=0
    #print("u=",U)
    for j in range(0,m):
        U[n,j]=40
        U[0,j]=0
    #print("u.=",U)
    coeff()
    for k1 in range(0,p1):
        for i in range(0,n+1):
            for j in range(0,m+1):
                U1[i,j]=U[i,j]
        for i in range(0,n+1):
            for j in range(0,m+1):
                U2[i,j]=U[i,j]
        #print("u2=",U2)
        gauss()
        for i in range(0,n+1):
            for j in range(0,m+1):
                U[i,j]=(alp)*U2[i,j]+(1-alp)*U[i,j]
        for i in range(1,n+1):
            U[i,m]=U[i,m-1]
        #print("update",U)
        
        max1=0
        flag=0
        for i in range(1,n):
            for j in range(1,m):
                max2=abs(U[i,j]-U1[i,j])
                if(max2>max1):
                    max1=max2
        R[k1]=max1
        R1[k1]=k1+1
        #print("R",R)
        #print(U)
        if(max1<eps):
            print("max1=",max1)
            flag=1
            #itmax=k1
            break
    print("iterations=",k1)
    plt.figure(figsize=(8,8))
    plt.plot(R1[0:k1],R[0:k1],color="red")
    plt.xlabel("iterations")
    plt.ylabel("residue")
    plt.ylim([-0.09,6])
    plt.xlim([-0.1,4000])
    plt.title("gauss siedal grid_size60x60")
    plt.show()
        
    if flag==1:
        print("solution converge ")
    else:
        print("diverge")
        
    #print(U)
    #k1=range(0,100)
    for i in range(0,n+1):
        r1=r[i]
        print ('r =',r1)
        print("theta\tu_numeric")
        for j in range(0,m+1):
            #if(i)
            #plt.plot()
            #u_exact=((300.0/pi)*((sin(pi)*x[i])*(sinh(pi)*y[j])))/sinh(pi)
            print('{0:1f} {1:2f}'.format(theta[j], U[i,j]))
            #plt.plot(theta)
    for i in range(0,n+1):
        #print("r=",r[i])
        for j in range(0,m+1):
            if(i==1.0):
                plt.scatter(theta[j],U[i,j])
                plt.xlabel("theta")
                plt.ylabel("u_numerical")
    plt.show()
main()
```
