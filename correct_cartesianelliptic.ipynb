```python
def coeff():
    for i in range(1,n):
        for j in range(1,m):
            A[i,j]=2.0/(h*h)+2.0/(k*k)
            B[i,j]=1.0/(h*h)
            C[i,j]=1.0/(h*h)
            D[i,j]=1.0/(k*k)
            E[i,j]=1.0/(k*k)
```

```python
def gauss():
    for k in range(1,p):
        for i in range(1,n):
            for j in range(1,m):
                U[i,j]=(B[i,j]*U[i-1,j]+C[i,j]*U[i+1,j]+D[i,j]*U[i,j-1]+E[i,j]*U[i,j+1])/A[i,j]
```
