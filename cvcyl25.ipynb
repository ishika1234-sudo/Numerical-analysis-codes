```python
def coeff():
    for i in range(1,n):
        for j in range(1,m):
            aP[i,j]=((r[i]+r[i+1])/2)*k/(r[i]*h)+((r[i]+r[i-1])/2)*k/(r[i]*h)+(h/k)+(h/k)
            aE[i,j]=((r[i]+r[i+1])/2)*k/(r[i]*h)
            aW[i,j]=((r[i]+r[i-1])/2)*k/(r[i]*h)                                                                                                                                                                                                                                                                                                                                                                                                                              
            aN[i,j]=h/k
            aS[i,j]=h/k
```

```python
def gauss():
    for k in range(1,p):
        for i in range(1,n):
            for j in range(1,m):
                U[i,j]=(aW[i,j]*U[i-1,j]+aE[i,j]*U[i+1,j]+aS[i,j]*U[i,j-1]+aN[i,j]*U[i,j+1])/aP[i,j]
```

```python
def main():
    import numpy as np
    from numpy import sin , pi, sinh
    global aE,aW,aN,aS,aP,U,U1,n,m,p,p1,h,k,r
    import seaborn as sns
    from numpy import sin , pi, sinh
    import matplotlib.pyplot as plt
    from mpl_toolkits import mplot3d
    n=20
    m=20
    p=3
    r=2.0
    z=4
    h=r/n
    print("h=",h)#delta r
    k=z/m
    print("k=",k)#delta z
    r0=0
    z0=0
    p1=1000
    eps=0.001
    R=np.zeros(p1)
    R1=np.zeros(p1)
    aP=np.zeros((n,m))
    aE=np.zeros((n,m))
    aW=np.zeros((n,m))
    aN=np.zeros((n,m))
    aS=np.zeros((n,m))
    U=np.zeros((n+1,m+1))
    r=np.zeros(n+1) 
    z=np.zeros(m+1)
    U1=np.zeros((n+1,m+1))
    R=np.zeros(p1)
    r[0]=r0
    for i in range(1,n+1):
        r[i]=r[i-1]+h
    print("r=",r)
    z[0]=z0
    for j in range(1,m+1):
        z[j]=z[j-1]+k
    print("z=",z)    
    
    for i in range(0,n+1):
        U[i,0]=0
        U[i,m]=50
    for j in range(0,m+1):
        U[n,j]=0
    #print("u=",U)
    #for j in range(0,m+1):
        #U[n,j]=50
    coeff()
    #outer iteration loop
    for k2 in range(0,p1):
        for i in range(0,n+1):
            for j in range(0,m+1):
                U1[i,j]=U[i,j] #storing the values to check convergence
        gauss()
        max1=0
        flag=0
        for i in range(1,n):
            for j in range(1,m):
                max2=abs(U[i,j]-U1[i,j])
                if(max2>max1):
                    max1=max2
        R[k2]=max1 #storing the residues in an array
        R1[k2]=k2+1 #storing the iterations in an array
        if(max1<eps):
            print("max1=",max1)
            flag=1
            break
    print("iteration=",k2)
    #plot of residue vs iterations
    plt.figure(figsize=(8,8))
    plt.plot(R1[0:k2],R[0:k2],color="red")
    plt.xlabel("iterations")
    plt.ylabel("residue")
    plt.ylim([-0.06,15])
    plt.xlim([0,200])
    plt.title("control volume 20x20")
    plt.show()
    if flag==1:
        print("solution converge ")
    else:
        print("diverge")
    print(U)
   
    for i in range(0,n+1):
        r1=r[i]
        print ('r =',r1)
        print("z\tu_numeric")
        for j in range(0,m+1):
            print('{0:1f} {1:2f}'.format(z[j], U[i,j]))
    #numerical plot
    for i in range(0,n+1):
        for j in range(0,m+1):
            if(i==1.0):
                plt.scatter(z[j],U[i,j])
    plt.show()
main()
```
